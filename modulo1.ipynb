### **Matching learning**

Machine Learning = aprendizaje autom√°tico, es una rama de IA que ense√±a a los ordenadores a aprender de los datos sin q los programes directamente.
En lugar de decirle paso a paso ‚Äúsi pasa esto, haz esto‚Äù,
le das muchos ejemplos, y el ordenador aprende por s√≠ mismo a encontrar patrones y tomar decisiones.

* CLASIFICACI√ìN NO-SUPERVISADA - yo no tengo soluciones, solo tengo datos y tengo q descubrir q tienen esos datos en comun.  
* CLASIFICACI√ìN SUPERVISADA - aqui tenemos un concepto matematico: la distancia, y pues tenemos un calculo para llegar a la sol.


**Clasificacion supervisada**
1.   Se en las dimensiones donde hay la info
2.   Se las soluciones
3.  Se cuantos clusters hay


* Clusters son grupos de datos que se parecen entre s√≠, creados autom√°ticamente por un algoritmo como K-Means.


* El m√©todo del codo (elbow method) es una t√©cnica muy usada en K-Means para elegir cu√°ntos clusters (grupos) usar en tus datos. La idea b√°sica:
* K-Means necesita que le digas cu√°ntos grupos (k) quieres crear.
Pero, ¬øc√≥mo saber si deben ser 2, 3 o 10 grupos? Ah√≠ entra el m√©todo del codo.
* haces su inercia (kmeans.inertia_), que mide
qu√© tan ‚Äújuntos‚Äù est√°n los puntos dentro de cada cluster.
*  Cuanto m√°s baja, mejor.
* un gr√°fico: Eje X ‚Üí n√∫mero de clusters (k) Eje Y ‚Üí inercia
* Al principio, cuando aumentas k, la inercia baja r√°pidamente. Luego llega un punto donde deja de mejorar tanto: ese punto forma una especie de ‚Äúcodo‚Äù en la gr√°fica.  El valor de k en el ‚Äúcodo‚Äù es el n√∫mero √≥ptimo de clusters: a√±adir m√°s grupos a partir de ah√≠ ya no mejora mucho el resultado.


**numero de clusers:**
* codo
* Silhouette Score --> Mide qu√© tan bien separados est√°n los clusters.
Va de -1 a 1 ‚Üí cuanto m√°s cerca de 1, mejor definidos est√°n los grupos.
Se elige el k con el mayor valor medio de silueta.
üëâ Eval√∫a cohesi√≥n interna y separaci√≥n externa.
* ch --> Calinski‚ÄìHarabasz (CH index): Compara la dispersi√≥n entre clusters con la dispersi√≥n interna.Cuanto mayor el valor, mejor la separaci√≥n entre clusters. üëâ Bueno para grandes vol√∫menes de datos.
* db --> Davies‚ÄìBouldin (DB index): Eval√∫a la similitud media entre clusters. Cuanto menor el √≠ndice, mejor el resultado. üëâ Busca clusters compactos y bien separados.


**algoritmos**
* kmeans (q es muy largo, pq calcula centro y luego calcula la distancia hasta cada punto)
* minibatchkmeans es mas rapido q kmeans, aunque hace mas error, pero es mas rapido para mas cantidades, usa muestras peque√±as (batches) en cada iteraci√≥n.
* MeanShift - el coge unos puntos y busca los mas densos y los usa para calcular. No requiere definir k. Busca regiones de alta densidad de puntos y define sus centros. Detecta autom√°ticamente el n√∫mero de clusters. Computacionalmente pesado con muchos datos.

üìà C√≥mo interpretar el resultado del **from sklearn.metrics import silhouette_score**: Valor del score	Interpretaci√≥n
* üü© 0.7 a 1.0	Clusters muy bien separados
* üü® 0.5 a 0.7	Buena estructura, razonable
* üüß 0.25 a 0.5	Estructura d√©bil o mezclada
* üü• < 0.25	Clustering malo o sin separaci√≥n clara

* Eje X (horizontal) ‚Üí n√∫mero de clusters que probaste (1, 2, 3, ...)
* Eje Y (vertical) ‚Üí la inercia: qu√© tan lejos est√°n los puntos de su centro (una medida del error interno del modelo)

üìâ C√≥mo se interpreta:
* Ô∏èAl principio (1 o 2 clusters), la inercia es muy alta üëâ los grupos son demasiado grandes y los puntos est√°n lejos del centro.
* Luego la inercia baja r√°pido üëâ significa que a√±adir m√°s clusters mejora mucho la agrupaci√≥n.
*  Pero llega un punto donde la curva deja de bajar tanto, y se ‚Äúdobla‚Äù formando una especie de codo.A partir de ah√≠, agregar m√°s clusters ya no mejora mucho el resultado = Ese ‚Äúcodo‚Äù es el n√∫mero √≥ptimo de clusters

*  En este gr√°fico, se ve m√°s o menos as√≠: De 1 a 3 clusters, la l√≠nea baja muy r√°pido y a partir de 3 o 4, la mejora se suaviza (ya casi no baja)
üëâ Por tanto, el n√∫mero ideal de clusters ser√≠a 3, que coincide perfectamente con las 3 especies del dataset Iris üå∏ (setosa, versicolor, virginica).
